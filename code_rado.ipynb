```python
import matplotlib.pyplot as plt
!pip install xlrd

import pandas as pd
url = ("https://www.sharkattackfile.net/spreadsheets/GSAF5.xls")
sharkattacks_data = pd.read_excel(url)
display(sharkattacks_data.head().tail())
```

```python
# NOW LET'S SEE THE NUMBER OF PROVOKED SHARK ATTACKS BY ACTIVITY

provoked_df = sharkattacks_data[
    sharkattacks_data["type"].astype(str).str.strip().str.lower() == "provoked"
]

activity_totals = (
    provoked_df.groupby("activity")
    .size()
    .reset_index(name="total")
    .sort_values("total", ascending=False)
    .reset_index(drop=True)
    .head(11)
)

activity_totals["activity"] = activity_totals["activity"].replace("Unknown", "unknown")

plt.figure(figsize=(10, 5))
plt.bar(activity_totals["activity"], activity_totals["total"])
plt.title("Total Provoked Shark Attacks by Activity")
plt.xlabel("Activity")
plt.ylabel("Total Attacks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# NOW LET'S SEE THE NUMBER OF UNPROVOKED SHARK ATTACKS BY ACTIVITY

unprovoked_totals = (
    sharkattacks_data[
        sharkattacks_data["type"].astype(str).str.strip().str.lower() == "unprovoked"
    ]
    .groupby("activity")
    .size()
    .reset_index(name="total")
    .sort_values("total", ascending=False)
    .reset_index(drop=True)
    .head(11)
)

unprovoked_totals["activity"] = unprovoked_totals["activity"].replace("Unknown", "unknown")

plt.figure(figsize=(10, 5))
plt.bar(unprovoked_totals["activity"], unprovoked_totals["total"])
plt.title("Total Unprovoked Shark Attacks by Activity")
plt.xlabel("Activity")
plt.ylabel("Total Attacks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
